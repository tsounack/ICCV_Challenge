```python
import numpy as np
import pandas as pd
```

```python
def read_hyp_file(path):
    data = []
    with open(path, 'r') as file:
        lines = file.readlines()
        sublist = []
        for line in lines:
            line = line.strip()
            
            if line.startswith('['):
                sublist = [float(value) for value in line[1:-1].split()]
            elif line.endswith(']'):
                sublist += [float(value) for value in line[:-1].split()]
                data.append(sublist)
            else:
                sublist += [float(value) for value in line.split()]
    return data
```

```python
# Read the labels from the labels.tok file
with open('labels.tok', 'r') as labels_file:
    labels = labels_file.read().splitlines()
labels1 = labels[1:] # Remove the first line

with open('labels.tok', 'r') as labels_file:
    labels = labels_file.read().splitlines()
labels2 = labels[1:] # Remove the first line

with open('labels3.tok', 'r') as labels_file:
    labels = labels_file.read().splitlines()
labels3 = labels[1:] # Remove the first line
```

```python
data1 = read_hyp_file('hyps.txt')
data2 = read_hyp_file('hyps2.txt')
data3 = read_hyp_file('hyps3.txt')

assert len(data1) == len(data2)
assert len(data1) == len(data3)
```

```python
import csv

def count_lines(csv_file):
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        line_count = sum(1 for row in reader)
    return line_count

csv_file = 'test.csv'
num_lines = count_lines(csv_file)
print(f"The CSV file contains {num_lines} lines.")

assert len(data1) == num_lines - 1
```

```python
df_predictions1 = pd.DataFrame(data1, columns=labels1)
df_predictions2 = pd.DataFrame(data2, columns=labels2)
df_predictions3 = pd.DataFrame(data3, columns=labels3)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df_predictions1.values)
df_predictions1 = pd.DataFrame(normalized_data, columns=df_predictions1.columns)

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df_predictions2.values)
df_predictions2 = pd.DataFrame(normalized_data, columns=df_predictions2.columns)

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df_predictions3.values)
df_predictions3 = pd.DataFrame(normalized_data, columns=df_predictions3.columns)
```

```python
threshold = 0.98989898989899
```

```python
df_predictions1.loc[df_predictions1['No positive label'] > threshold, :] = 0
df_predictions1 = df_predictions1.drop('No positive label', axis=1)

df_predictions2.loc[df_predictions2['No positive label'] > threshold, :] = 0
df_predictions2 = df_predictions2.drop('No positive label', axis=1)

df_predictions3.loc[df_predictions3['No positive label'] > threshold, :] = 0
df_predictions3 = df_predictions3.drop('No positive label', axis=1)
```

```python
csv_file_path = 'test.csv'
df_csv = pd.read_csv(csv_file_path)
first_column = df_csv.iloc[:, 0]
```

```python
df_predictions1.insert(0, 'dicom_id', first_column)
df_predictions2.insert(0, 'dicom_id', first_column)
df_predictions3.insert(0, 'dicom_id', first_column)
```

```python
for column in df_predictions1.columns[1:]:
    df_predictions1.loc[df_predictions1[column] >= 1, column] = 1
    df_predictions1.loc[df_predictions1[column] <= 0, column] = 0

for column in df_predictions2.columns[1:]:
    df_predictions2.loc[df_predictions2[column] >= 1, column] = 1
    df_predictions2.loc[df_predictions2[column] <= 0, column] = 0

for column in df_predictions3.columns[1:]:
    df_predictions3.loc[df_predictions3[column] >= 1, column] = 1
    df_predictions3.loc[df_predictions3[column] <= 0, column] = 0
```

```python
first_column = df_predictions1.iloc[:, 0:1]
sorted_columns = sorted(df_predictions1.columns[1:])
df_predictions1 = pd.concat([first_column, df_predictions1[sorted_columns]], axis=1)

first_column = df_predictions2.iloc[:, 0:1]
sorted_columns = sorted(df_predictions2.columns[1:])
df_predictions2 = pd.concat([first_column, df_predictions2[sorted_columns]], axis=1)

first_column = df_predictions3.iloc[:, 0:1]
sorted_columns = sorted(df_predictions3.columns[1:])
df_predictions3 = pd.concat([first_column, df_predictions3[sorted_columns]], axis=1)
```

```python
df_predictions1_numerical = df_predictions1.iloc[:, 1:].values
df_predictions2_numerical = df_predictions2.iloc[:, 1:].values
df_predictions3_numerical = df_predictions3.iloc[:, 1:].values

weight_df_predictions3 = 0.6
weight_df_predictions2 = 0.3
weight_df_predictions1 = 0.1

weighted_avg_numerical = (
    weight_df_predictions1 * df_predictions1_numerical
    + weight_df_predictions2 * df_predictions2_numerical
    + weight_df_predictions3 * df_predictions3_numerical
) / (weight_df_predictions1 + weight_df_predictions2 + weight_df_predictions3)

columns = df_predictions1.columns.tolist()
final_data = np.hstack((first_column.values.reshape(-1, 1), weighted_avg_numerical))
df_predictions_final = pd.DataFrame(final_data, columns=columns)
```
