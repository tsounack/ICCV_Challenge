```python
# Read the JSON file
with open(path2 + 'ckpt6_metrics.txt') as file:
    data2 = json.load(file)

# Extract the "ap" list from the JSON data
ap_list2 = data2['scores']['ap']

# Create a DataFrame
df2 = pd.DataFrame({'Label': labels2, 'AP': ap_list2})
df2 = df2.drop(df2[df2['Label'] == 'No positive label'].index)
df2.sort_values(by='Label', inplace=True)
df2.reset_index(drop=True, inplace=True)
df2 = df2.append({'Label': 'Average', 'AP': df2['AP'].mean()}, ignore_index=True)
```
